```python
with open('piwikresponse.json', 'r') as file: 
    data = json.load(file)

flattened_data = [item for sublist in data['data'] for item in sublist]

with open('flattened_data.json', 'w') as file:
    json.dump({"data": flattened_data}, file)
```

```python
import json

# Provided JSON object


# Flatten the data
flattened_data = []
for entry in data['data']:
    flattened_entry = {}
    for i, column in enumerate(data['meta']['columns']):
        if column == "referrer_type":
            flattened_entry["referrer_id"] = entry[i][0]
            flattened_entry["referrer_name"] = entry[i][1]
        else:
            flattened_entry[column] = entry[i]
    flattened_data.append(flattened_entry)

# Create the flattened JSON object
flattened_json = {
    "data": flattened_data,
    "meta": data['meta']
}

with open('flattened_data_transformed.json', 'w') as outfile:
    json.dump(flattened_json, outfile, indent=4)
```
